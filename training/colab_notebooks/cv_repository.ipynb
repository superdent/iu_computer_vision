```python
# Setup: Drive mounten, SSH-Key laden, Git-Config setzen
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os, subprocess, textwrap, pathlib
os.makedirs(os.path.expanduser("~/.ssh"), exist_ok=True)

# Privaten Schlüssel aus Drive kopieren
src = "/content/drive/MyDrive/ssh/git_googlecolab"  # dein Key
dst = os.path.expanduser("~/.ssh/id_ed25519")
subprocess.run(["/bin/bash","-lc", f"cp '{src}' '{dst}' && chmod 600 '{dst}'"], check=True)

# SSH-Config + known_hosts
cfg = textwrap.dedent("""\
Host github.com
  HostName github.com
  User git
  IdentityFile ~/.ssh/id_ed25519
  IdentitiesOnly yes
""")
pathlib.Path(os.path.expanduser("~/.ssh/config")).write_text(cfg)
subprocess.run(["/bin/bash","-lc","ssh-keyscan github.com >> ~/.ssh/known_hosts"], check=True)

# Git-Commit-Identität
subprocess.run(["git","config","--global","user.name","superdent"], check=True)
subprocess.run(["git","config","--global","user.email","ruediger.schlenter@gmx.de"], check=True)

!git config --global core.filemode false
!git config --global core.autocrlf input
!git config --global safe.directory '*'
```
