```python
from google.colab import drive
import shutil
import zipfile
import os

# Google Drive einbinden
drive.mount('/content/drive')

# Pfade definieren
src_zip = '/content/drive/MyDrive/gtsrb/yolo_training.zip'
dst_dir = '/content'

# Zielverzeichnis erstellen (falls nicht vorhanden)
os.makedirs(dst_dir, exist_ok=True)

# ZIP entpacken
with zipfile.ZipFile(src_zip, 'r') as zip_ref:
    zip_ref.extractall(dst_dir)

print("Datei erfolgreich entpackt nach:", dst_dir)
```

```python
count = sum(len(files) for _, _, files in os.walk('/content/yolo_training'))
print(f"Anzahl entpackter Dateien: {count}")
```
